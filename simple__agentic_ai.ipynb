```python
import datetime
import re

class SimpleAgent:
    """
    A tiny 'agentic' AI style program:
    - Takes user text
    - Decides which TOOL to use
    - Returns the tool's output
    """

    def __init__(self):
        # You could register more tools here
        self.tools = {
            "calculator": self._calculator_tool,
            "time": self._time_tool,
            "echo": self._echo_tool,
        }

    # ---------------------------
    # MAIN AGENT LOGIC
    # ---------------------------
    def handle_input(self, user_text: str) -> str:
        """
        Decide what to do with the user text.
        This is like a very small 'planner + router'.
        """

        text = user_text.lower()

        # 1ï¸âƒ£ If the user wants to calculate something
        if any(word in text for word in ["add", "sum", "plus", "+", "minus", "subtract", "multiply", "divide", "*", "/"]):
            return self.tools["calculator"](user_text)

        # 2ï¸âƒ£ If the user asks for time
        if "time" in text or "current time" in text:
            return self.tools["time"](user_text)

        # 3ï¸âƒ£ Otherwise, just echo / reply back
        return self.tools["echo"](user_text)

    # ---------------------------
    # TOOLS
    # ---------------------------

    def _calculator_tool(self, user_text: str) -> str:
        """
        Very simple calculator tool.
        Supports things like:
        - 'add 2 and 3'
        - '2 + 3'
        - '10 - 4'
        - '6 * 7'
        - '8 / 2'
        """

        # Try to extract an expression like "2 + 3" using regex
        # If user wrote "add 2 and 3" we convert it roughly to "2 + 3"
        text = user_text.lower()

        # Replace words with symbols
        text = text.replace("add", " ")
        text = text.replace("plus", " + ")
        text = text.replace("minus", " - ")
        text = text.replace("subtract", " - ")
        text = text.replace("into", " * ")
        text = text.replace("multiply", " * ")
        text = text.replace("x", " * ")
        text = text.replace("divide by", " / ")
        text = text.replace("divided by", " / ")

        # Pick something that looks like "number op number"
        match = re.search(r"(-?\d+(\.\d+)?)\s*([\+\-\*/])\s*(-?\d+(\.\d+)?)", text)
        if not match:
            return "I couldn't understand the math expression. Try like: 2 + 3 or add 10 and 20."

        a = float(match.group(1))
        op = match.group(3)
        b = float(match.group(4))

        try:
            if op == "+":
                result = a + b
            elif op == "-":
                result = a - b
            elif op == "*":
                result = a * b
            elif op == "/":
                if b == 0:
                    return "Division by zero is not allowed."
                result = a / b
            else:
                return "Unknown operator."

            # If result is int-like, cast to int for clean output
            if result.is_integer():
                result = int(result)

            return f"ðŸ§® Calculator result: {a} {op} {b} = {result}"
        except Exception as e:
            return f"Error while calculating: {e}"

    def _time_tool(self, _user_text: str) -> str:
        """
        Tool that returns the current time.
        """
        now = datetime.datetime.now()
        return f"â° Current time is: {now.strftime('%Y-%m-%d %H:%M:%S')}"

    def _echo_tool(self, user_text: str) -> str:
        """
        Fallback tool â€“ just responds politely.
        """
        return f"ðŸ¤– I received your message: '{user_text}'\nRight now I can calculate, tell time, or just chat."

# ---------------------------
# SIMPLE CLI LOOP
# ---------------------------

if __name__ == "__main__":
    agent = SimpleAgent()

    print("=== Simple Agentic-Style AI Demo ===")
    print("Type something (e.g., 'add 2 and 3', 'what is the time?', or 'hello').")
    print("Type 'exit' or 'quit' to stop.\n")

    while True:
        user_input = input("You: ")

        if user_input.lower() in ["exit", "quit", "bye"]:
            print("Agent: Bye! ðŸ‘‹")
            break

        response = agent.handle_input(user_input)
        print("Agent:", response)
        print()
```
